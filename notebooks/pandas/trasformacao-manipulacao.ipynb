# Transformação e Manipulação de Dados

## Const and imports

```python
import re
import pandas as pd
import numpy as np
from string import punctuation
```

## Aula

```python
CSV_1 = 'https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_hospedagem.json'
CSV_2 = 'https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/moveis_disponiveis.json'
```

```python
df_hospedagem = pd.read_json(CSV_1)
df_hospedagem.head()
```

```python
df_hospedagem = pd.json_normalize(df_hospedagem['info_moveis'])
```

```python
df_hospedagem.head()
```

```python
df_hospedagem = df_hospedagem.explode(list(df_hospedagem.columns)[3:])
```

```python
df_hospedagem.reset_index(inplace=True, drop=True)
```

```python
df_hospedagem.head()
```

```python
df_hospedagem.info()
```

```python
cols_to_int64 = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
df_hospedagem[cols_to_int64] = df_hospedagem[cols_to_int64].astype(np.int64)
```

```python
df_hospedagem['avaliacao_geral'] = df_hospedagem['avaliacao_geral'].astype(np.float64)
```

```python
df_hospedagem['preco'] = df_hospedagem['preco'].apply(lambda el: el.replace(r'$', '').replace(',', ''))
```

```python
df_hospedagem['preco'] = df_hospedagem['preco'].astype(np.float64)
```

```python
df_hospedagem[['taxa_deposito', 'taxa_limpeza']] = df_hospedagem[['taxa_deposito', 'taxa_limpeza']].applymap(lambda el: el.replace('$', '').replace(',', ''))
```

```python
df_hospedagem[['taxa_deposito', 'taxa_limpeza']] = df_hospedagem[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

```python
df_hospedagem['max_hospedes'] = df_hospedagem['max_hospedes'].astype(np.int64)
```

```python
df_hospedagem[['descricao_local', 'descricao_vizinhanca']] = df_hospedagem[['descricao_local', 'descricao_vizinhanca']].applymap(lambda el: el.lower())
```

```python
df_hospedagem['descricao_local'] = df_hospedagem['descricao_local'].str.replace(r'[^a-zA-Z0-9-\']', ' ', regex=True)
```

```python
df_hospedagem['descricao_local'] = df_hospedagem['descricao_local'].replace(r'(?<!\w)-(?!\w)', '', regex=True)
```

```python
df_hospedagem['descricao_local'].str.split(' ')
```

```python
df_hospedagem['comodidades'] = df_hospedagem['comodidades'].str.replace(r'\{|}|"', '', regex=True).str.split(',')
```

## Atividade 1 - Vendas de Clientes

```python
CSV_3 = BASE_PATH + '/dados_vendas_clientes.json'
```

```python
df_vendas_clientes = pd.read_json(CSV_3)
df_vendas_clientes.head()
```

```python
df_vendas_clientes = pd.json_normalize(df_vendas_clientes['dados_vendas'])
```

```python
df_vendas_clientes.head()
```

```python
df_vendas_clientes = df_vendas_clientes.explode(list(df_vendas_clientes.columns)[1:]).reset_index(drop=True)
```

```python
df_vendas_clientes['Valor da compra'] = df_vendas_clientes['Valor da compra'].apply(lambda el: el.replace('R$', '').replace(',', '.'))
```

```python
df_vendas_clientes['Valor da compra'] = df_vendas_clientes['Valor da compra'].astype(np.float64)
```

```python
df_vendas_clientes['Data de venda'] = df_vendas_clientes['Data de venda'].apply(lambda el: pd.to_datetime(el, format='%d/%m/%Y', errors='ignore'))
```

```python
df_vendas_clientes['Cliente'] = df_vendas_clientes['Cliente'].str.lower().replace(
    r'[{exp}]'.format(exp='|'.join([*punctuation])),
    '',
    regex=True).replace(r'[0-9]', '', regex=True)
```

```python
df_vendas_clientes_subset = df_vendas_clientes.groupby('Cliente').sum()
```

```python
df_vendas_clientes_subset.sort_values(by='Valor da compra', ascending=False)
```

## Atividade 2 - Locação de Imóveis

```python
CSV_4 = BASE_PATH + '/dados_locacao_imoveis.json'
```

```python
df_locacao_imoveis = pd.read_json(CSV_4)
df_locacao_imoveis.head()
```

```python
df_locacao_imoveis = pd.json_normalize(df_locacao_imoveis['dados_locacao'])
```

```python
df_locacao_imoveis.head()
```

```python
df_locacao_imoveis = df_locacao_imoveis.explode(list(df_locacao_imoveis.columns)[1:]).reset_index(drop=True)
```

```python
cols_to_date = ['datas_combinadas_pagamento', 'datas_de_pagamento']
df_locacao_imoveis[cols_to_date] = df_locacao_imoveis[cols_to_date].applymap(lambda el: pd.to_datetime(el, format='%d/%m/%Y', errors='ignore'))
```

```python
df_locacao_imoveis['valor_aluguel'] = df_locacao_imoveis['valor_aluguel'].apply(lambda el: re.sub(r'[^\d,]+', '', el).replace(',', '.'))
```

```python
df_locacao_imoveis['valor_aluguel'] = df_locacao_imoveis['valor_aluguel'].astype(np.float64)
```

```python
df_locacao_imoveis['apartamento'] = df_locacao_imoveis['apartamento'].str.split().apply(lambda el: el[0])
```

```python
df_locacao_imoveis['atraso_dias'] = (df_locacao_imoveis['datas_de_pagamento'] - df_locacao_imoveis['datas_combinadas_pagamento']).dt.days
```

```python
df_locacao_imoveis.groupby('apartamento').sum().sort_values('atraso_dias', ascending=False)
```
