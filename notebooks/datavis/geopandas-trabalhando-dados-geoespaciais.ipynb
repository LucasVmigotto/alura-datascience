## Aula 2 - Trabalhando com sistemas de coordenadas

```python
import pandas as pd
```

```python
df = pd.read_csv(CSV_URL, sep='\t')

df.head()
```

```python
list(zip(df['Longitude'], df['Latitude']))
```

```python
from shapely import Point
```

```python
pontos_geometricos = [Point(x) for x in zip(df['Longitude'], df['Latitude'])]

pontos_geometricos
```

### Entendendo o sistema de coordenadas

#### CRS - Coordinate Reference System

* Projeção `proj`: O tipo da projeção que é feita (cilíndica, azimutal, cônica...)
    * `latlong`
    * `utm` (Universal Transverse Mercator)
        * Favorecimento a diminuição de distorções em situações de cálculos de distâncias
        * É preciso informar, dentre as 60 zonas, aquela que a projeção se localiza
    * `tmerc`

* Elipsóide `ellps`: A construção da elipse
    * `WGS84`
    * `GRS80`
    * `CRLK80`

* `datum`: Define o sistema de eixos e projeção central
    * `WGS84`
    * `SIRGAS 2000` (Sistema adotado oficialmente pelo Brasil)

```python
csr = {'proj': 'latlong',
       'ellps': 'WGS84',
       'datum': 'WGS84',
       'no_defs': True}

geo_dados = gpd.GeoDataFrame(df, crs=csr, geometry=pontos_geometricos)

geo_dados.head()
```

```python
geo_dados.plot(alpha=.2)
```

```python
rj.crs
```

```python
geo_dados.crs
```

```python
# Qual a zona UTM em que o município do Rio de Janeiro está inserido?
# R: Zona 23S

# Qual o EPSG para a zona UTM 23S utilizando o SIRGAS 2000?
# R: 31983
# Referêcias
#      * https://spatialreference.org/
#      * https://spatialreference.org/ref/epsg/31983/proj4/

CRS_CODE = '+proj=utm +zone=23 +south +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=km +no_defs '

rj = rj.to_crs(CRS_CODE)

rj.crs
```

```python
geo_dados = geo_dados.to_crs(CRS_CODE)

geo_dados.crs
```

```python
rj.plot()
```

## Aula 4 - Calculando distâncias

```python
metro_distancias = df_metro.distance(geo_dados.iloc[0].geometry)

metro_distancias.head()
```

```python
geo_dados['dist_metro'] = geo_dados['geometry'].apply(lambda x: df_metro.distance(x).min())

geo_dados.head()
```

```python
geo_dados.corr()
```

```python
import seaborn as sns

sns.heatmap(geo_dados.corr(),
             annot=True,
             fmt='.2f')
```

```python
df_trem = gpd.read_file(RJ_TREM)

df_trem.head()
```

```python
df_trem = df_trem.to_crs(CRS_CODE)

df_trem.crs
```

```python
df_trem = df_trem[df_trem.within(rj.iloc[0].geometry)]
df_trem.head()
```

```python
df_brt = gpd.read_file(RJ_BRT)

df_brt.head()
```

```python
df_brt = df_brt.to_crs(CRS_CODE)

df_brt.crs
```

```python
df_brt = df_brt[df_brt.within(rj.iloc[0].geometry)]
df_brt.head()
```

```python
base = rj.plot(color='white', edgecolor='black')
geo_dados.plot(ax=base, alpha=.2)
df_metro.plot(ax=base, color='red', markersize=10)
df_trem.plot(ax=base, color='green', markersize=10)
df_brt.plot(ax=base, color='pink', markersize=10)
```

```python
transporte = pd.concat([df_trem.geometry,
                        df_metro.geometry,
                        df_brt.geometry],
                       ignore_index=True)

transporte.head()
```

```python
geo_dados['dist_transporte'] = geo_dados['geometry'].apply(lambda x: transporte.distance(x).min())

geo_dados.head()
```

```python
sns.heatmap(geo_dados.corr(),
             annot=True,
             fmt='.2f')
```
